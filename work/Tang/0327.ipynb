```python
import pandas as pd

class Hedge(object):
    
    #########
    # TODOt
    # For now, don't have tolerance for nan data.
    #########
    
    def __init__(self, allstockdf, begindate = None, enddate = None):
        # Use AAPL dates to filter
        # preprossessing
        idxb = 0
        if begindate != None:
            idxb = np.where(allstockdf.Date==begindate)[0][0]
        if enddate != None:
            idxe = np.where(allstockdf.Date==enddate)[0][0]
            allstockdfsliced = allstockdf[idxb:idxe+1]
        else:
            allstockdfsliced = allstockdf[idxb:]
        self.allstockdf = allstockdf.iloc[allstockdfsliced['AAPL'].dropna().index.values]
        self.allstockdf = self.allstockdf.dropna(axis = 1)
    
```

```python
hedgeobj = Hedge(stock_df, begindate = '2010-01-04')
```
