```python
import numpy as np
import Quandl
import pandas as pd
from stockSelection import *
import pickle
```

```python
stock_universe = getStockUniverse()
stocks = list(stock_universe.Symbol)
```

```python
r = getStocks(stocks)
```

```python
# Construct dataframe from dict of stocks

stocks_df = pd.concat(r, axis=1)
stocks_df.columns = stocks_df.columns.droplevel(1)
```

```python
# Find which stocks have valid data for which years

valid_stocks = {}
for year in range(2005, 2016):
    # Get first and last three dates of year
    first = stocks_df.ix[stocks_df.index[stocks_df.index.year == year][:3]].notnull()
    last = stocks_df.ix[stocks_df.index[stocks_df.index.year == year][-3:]].notnull()
    
    # Get all stocks that are valid on one of the first three days of the year,
    # and at least one of the last three.
    fidx = first.index
    lidx = last.index
    valid_first = first.ix[fidx[0]] | first.ix[fidx[1]] | first.ix[fidx[2]]
    valid_last = last.ix[lidx[0]] | last.ix[lidx[1]] | last.ix[lidx[2]]
    # Get all valid stocks
    valid_stocks[year] = list(first.columns[valid_first | valid_last])
```

```python
# Read dataframe to csv

stocks_df.to_csv('../../data/all_stocks.csv')
# Uncomment to make sure everything worked
# test_df = pd.read_csv('../../data/all_stocks.csv')
```
